# Reading Data In Pandas


```python
##Read Header

print(df.columns) # it will display all the columns in the table 
#df.columns

#print(df.columns[2]) # Single specified column(according to indexing)
```

```python
## Read Each Columns
print(df['Name'])
#df['Name']
```

```python
##Read multiple columns in table

print(df[['Name','Type 1','Type 2']])
```

```python
#Read each row
##Single Row

print(df.iloc[4]) #it will display the 4th raw details 
```

```python
##Multiple Rows

print(df.iloc[4:8]) #it will display 4rows(4 to 7 th raw) rows in the table( depends on numbers specified in [])
```

```python
#Read specific location on table by index number 

print(df.iloc[2,1])
```

```python
#Read Particular Location on Table by specific name 
df.loc[df['Name']=='Charmander']
```

# Filtering Our Data

```python
#Single Columns Filtering 
df.loc[df['Type 1']== 'Grass']
```

```python
#Multi Columns Filtering 

df.loc[(df['Type 1'] == 'Grass') | (df['Type 2'] == 'Poison')]
```

```python
#Multi Columns Filtering with Greater number 

df.loc[(df['Type 1'] == 'Grass') | (df['Type 2'] == 'Poison') & (df['HP']>40)]
```

# Condition Changes

```python
#If  you want to change Rows names

df.loc[df['Type 1']== 'Grass','Type 1']='Fire'
df.head()
```

```python
#it will change the row values i,e Genereation,Legendry row values ,if the Total is more then 500 

df.loc[df['Total']>500,['Generation','Legendary']]='test'
df.head()
```

# Aggreagte Statistics

```python
#mean values of the Type 1 Columns
df.groupby(['Type 1']).mean()
```

```python
#mean values with ascending 
df.groupby(['Type 1']).mean().sort_values('Defense',ascending=False)
```

```python
# Sum

df.groupby(['Type 2']).sum()
```

```python
#count
df.groupby(['Type 2']).count()
```
